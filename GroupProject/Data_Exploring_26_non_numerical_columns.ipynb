```python
import pandas as pd

data = pd.read_csv("15_more_to_be_imputed_data.csv")
```

```python
data.head()
```

```python
data.shape
```

```python
non_float_columns_header = []

for column in data.columns:
    if data[column].dtype != "float64":
#         print(column)
        non_float_columns_header.append(column)
#         print(data[column].dtype)
```

```python
non_float_columns = data[non_float_columns_header]

non_float_columns.head()
```

```python
non_float_columns.shape
```

```python
non_float_columns.to_csv("non_float_data.csv", index=False)
```
